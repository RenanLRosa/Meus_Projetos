```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Carregar os dados
tabela = pd.read_csv('vendas.csv')

# Grfico de barras - Preo dos produtos
plt.figure(figsize=(8,5))
sns.barplot(x='Produto', y='Preco_Unitario', data=tabela, palette='viridis')
plt.xticks(rotation=45)  # Rotaciona os nomes para melhor leitura
plt.title('Preo dos Produtos')
plt.xlabel('Produto')
plt.ylabel('Preo Unitrio (R$)')
plt.show()

# Histograma - Distribuio dos preos
plt.figure(figsize=(8,5))
sns.histplot(tabela['Preco_Unitario'], bins=5, kde=True, color='blue')
plt.title('Distribuio dos Preos')
plt.xlabel('Preo Unitrio (R$)')
plt.ylabel('Frequncia')
plt.show()
```
