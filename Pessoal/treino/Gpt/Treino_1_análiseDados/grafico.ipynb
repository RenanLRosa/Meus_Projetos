```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Carregar os dados
tabela = pd.read_csv('vendas.csv')

# Gráfico de barras - Preço dos produtos
plt.figure(figsize=(8,5))
sns.barplot(x='Produto', y='Preco_Unitario', data=tabela, palette='viridis')
plt.xticks(rotation=45)  # Rotaciona os nomes para melhor leitura
plt.title('Preço dos Produtos')
plt.xlabel('Produto')
plt.ylabel('Preço Unitário (R$)')
plt.show()

# Histograma - Distribuição dos preços
plt.figure(figsize=(8,5))
sns.histplot(tabela['Preco_Unitario'], bins=5, kde=True, color='blue')
plt.title('Distribuição dos Preços')
plt.xlabel('Preço Unitário (R$)')
plt.ylabel('Frequência')
plt.show()
```
